## 🤔 Part 1：为什么需要一个建模框架？

上一课我们学会了 Ontology 的五大构件，但面对一个全新的业务场景时，你可能会困惑：

> ❓ "我该从哪里开始？先定义类还是先定义属性？"
> ❓ "怎么知道我的模型是不是完整的？"
> ❓ "有没有一个通用的步骤可以遵循？"

**OPERA 框架**就是为此而生的 —— 它把 Ontology 建模分解为 **5 个有序的步骤**，按顺序做就不会遗漏。

````python
# 🎭 OPERA 框架全景

OPERA = {
    "O": {
        "全称": "Object Types (对象类型)",
        "一句话": "你的领域中有哪些'东西'？",
        "对应构件": "类 (Class) + 实例 (Individual)",
        "类比": "Excel 的表名 + 行数据",
        "关键问题": "这个世界里有哪些'名词'？"
    },
    "P": {
        "全称": "Properties (属性)",
        "一句话": "每种'东西'有哪些特征？",
        "对应构件": "数据属性 (Data Property)",
        "类比": "Excel 的列名和列类型",
        "关键问题": "每种对象需要记录什么信息？"
    },
    "E": {
        "全称": "Events & Actions (事件与动作)",
        "一句话": "会发生什么？能做什么？",
        "对应构件": "方法 / 触发器",
        "类比": "业务流程中的'动词'",
        "关键问题": "这个世界里有哪些'动词'？"
    },
    "R": {
        "全称": "Relations (关系)",
        "一句话": "东西和东西之间怎么连接？",
        "对应构件": "对象属性 (Object Property)",
        "类比": "数据库的外键关联",
        "关键问题": "两个对象之间有什么语义联系？"
    },
    "A": {
        "全称": "Axioms (公理/规则)",
        "一句话": "必须遵守什么规则？",
        "对应构件": "约束 / 推理规则",
        "类比": "数据库约束 + 业务规则引擎",
        "关键问题": "什么是必须的？什么是不允许的？"
    }
}

print("🎭 O P E R A 万能建模框架")
print("=" * 65)
print()

for letter, info in OPERA.items():
    print(f"  {letter}  │  {info['全称']}")
    print(f"     │  💬 {info['一句话']}")
    print(f"     │  🔗 对应构件：{info['对应构件']}")
    print(f"     │  ❓ 关键问题：{info['关键问题']}")
    print(f"     │")

print("=" * 65)
print("\n💡 记忆技巧：OPERA = 歌剧。建模就像编排一部歌剧：")
print("   演员(O) + 外貌服装(P) + 剧情动作(E) + 人物关系(R) + 规矩(A)")
````

## 🔤 Part 2：逐个击破 OPERA

### O — Object Types（对象类型）

> **核心问题：这个领域里有哪些"名词"？**

**建模三步法：**
1. 列出所有核心名词（人、事、物、地点、文档...）
2. 建立层级关系（is-a 继承树）
3. 区分"类模板"和"具体实例"

**🏆 黄金法则：**
- 如果它代表一类事物 → 它是 **Class**
- 如果它是某类事物的一个具体个体 → 它是 **Individual**
- 宁可多分类，不要把不同概念混在一起

````python
# 🏷️ O — 对象类型建模实践

print("📝 场景：为一所「大学」建模\n")
print("第一步：头脑风暴——列出所有名词")
print("-" * 50)
nouns = ["学生", "教师", "课程", "教室", "学院", "学位", "成绩", "论文", "学期"]
for i, n in enumerate(nouns, 1):
    print(f"  {i}. {n}")

print("\n第二步：建立层级关系（is-a 树）")
print("-" * 50)
hierarchy = """
  Person (人)
  ├── Student (学生)
  │   ├── Undergraduate (本科生)
  │   └── Graduate (研究生)
  │       ├── MasterStudent (硕士)
  │       └── PhDStudent (博士)
  └── Teacher (教师)
      ├── Professor (教授)
      ├── AssocProfessor (副教授)
      └── Lecturer (讲师)

  AcademicUnit (教学单位)
  ├── College (学院)
  └── Department (系)

  AcademicItem (教学要素)
  ├── Course (课程)
  ├── Semester (学期)
  └── Degree (学位)
"""
print(hierarchy)

print("第三步：区分类和实例")
print("-" * 50)
print("  类(模板):  Student, Professor, Course")
print("  实例(个体): 张三(Student), 李教授(Professor), 高等数学(Course)")
print("\n💡 类是'蓝图'，实例是'按蓝图造出来的具体东西'")
````

### P — Properties（属性）

> **核心问题：每种对象需要记录什么信息？**

**建模要点：**
1. 为每个类列出需要记录的数据字段
2. 明确每个属性的**数据类型**（string / int / float / boolean / datetime / enum）
3. 标注哪些是**必填**的、哪些是可选的
4. 考虑是否有**默认值**

**🏆 黄金法则：**
- 属性属于某个类（有 domain）
- 属性的值有类型（有 range）
- 如果 A 的属性指向 B 这个对象 → 那是"关系"(R)，不是属性(P)

````python
# 📋 P — 属性建模实践

print("📋 为大学本体定义属性\n")

properties = {
    "Student": [
        ("student_id",  "string",   True,  "学号"),
        ("name",        "string",   True,  "姓名"),
        ("gender",      "enum[M,F]", True, "性别"),
        ("birth_date",  "date",     True,  "出生日期"),
        ("gpa",         "float",    False, "绩点，0.0-4.0"),
        ("enrolled",    "boolean",  True,  "是否在读"),
    ],
    "Course": [
        ("course_id",   "string",   True,  "课程编号"),
        ("title",       "string",   True,  "课程名称"),
        ("credits",     "integer",  True,  "学分，1-6"),
        ("max_students","integer",  False, "最大选课人数"),
        ("description", "string",   False, "课程描述"),
    ],
    "Teacher": [
        ("teacher_id",  "string",   True,  "工号"),
        ("name",        "string",   True,  "姓名"),
        ("title",       "enum[教授,副教授,讲师]", True, "职称"),
        ("research_area","string",  False, "研究方向"),
    ]
}

for cls, props in properties.items():
    print(f"  📦 {cls}")
    print(f"  {'属性名':<16} {'类型':<14} {'必填':<6} {'说明'}")
    print(f"  {'─'*58}")
    for name, dtype, required, desc in props:
        req = "✅" if required else "  "
        print(f"  {name:<16} {dtype:<14} {req:<6} {desc}")
    print()

print("💡 注意：'张三所在的学院'不是 Student 的属性，而是关系(R)——")
print("   因为它指向的是另一个对象(College)，不是一个简单的值。")
````

### E — Events & Actions（事件与动作）

> **核心问题：这个世界里会发生什么？能做什么？**

这是 OPERA 中最容易被忽视、但最有实战价值的一环。

- **Event（事件）** — 自动发生的触发型逻辑，如"当选课人数满 → 自动关闭选课"
- **Action（动作）** — 可以主动调用的操作，如"为学生注册选课"

**🏆 黄金法则：**
- 每个 Action 写清楚：**谁(Subject)** + **做什么(Verb)** + **对谁/什么(Object)**
- 每个 Event 写清楚：**当什么条件满足时** + **自动做什么**
- 想想 Palantir：Action 就是 Ontology 上可以"点击执行"的按钮

````python
# ⚡ E — 事件与动作建模实践

print("⚡ 为大学本体定义事件与动作\n")

actions = [
    {
        "名称": "Enroll (选课)",
        "主体": "Student",
        "对象": "Course",
        "前置条件": "学生在读 + 未满课人数 + 未重复选课",
        "后置效果": "创建 enrollment 关系 + 选课人数 +1",
        "失败场景": "课程已满 → 加入等待列表"
    },
    {
        "名称": "GradeSubmit (提交成绩)",
        "主体": "Teacher",
        "对象": "Student × Course → Grade",
        "前置条件": "教师是该课程的授课教师",
        "后置效果": "创建 Grade 记录 + 更新学生 GPA",
        "失败场景": "非授课教师 → 拒绝操作"
    },
    {
        "名称": "Graduate (毕业)",
        "主体": "Student",
        "对象": "Degree",
        "前置条件": "所有必修课已通过 + GPA ≥ 2.0 + 论文通过",
        "后置效果": "授予学位 + 更新 enrolled=False",
        "失败场景": "条件不满足 → 列出缺失要求"
    }
]

events = [
    {
        "名称": "CourseFull (课程满员)",
        "触发条件": "当选课人数 == max_students",
        "自动动作": "关闭选课入口 + 通知教务处"
    },
    {
        "名称": "GPA Alert (绩点预警)",
        "触发条件": "当学生 GPA < 1.5",
        "自动动作": "标记学业预警 + 通知辅导员"
    }
]

print("🎬 动作 (Actions)：")
print("=" * 60)
for a in actions:
    print(f"\n  📌 {a['名称']}")
    print(f"     主体 → 对象: {a['主体']} → {a['对象']}")
    print(f"     前置条件: {a['前置条件']}")
    print(f"     后置效果: {a['后置效果']}")
    print(f"     失败场景: {a['失败场景']}")

print(f"\n\n⚡ 事件 (Events)：")
print("=" * 60)
for e in events:
    print(f"\n  🔔 {e['名称']}")
    print(f"     触发条件: {e['触发条件']}")
    print(f"     自动动作: {e['自动动作']}")
````

### R — Relations（关系）

> **核心问题：对象之间有什么语义联系？**

**常见关系类型：**
- `is-a` — 继承关系（Dog is-a Animal）
- `part-of` — 组成关系（Department part-of College）
- 业务关系 — `teaches`, `enrolled_in`, `supervises` 等

**🏆 黄金法则：**
- 关系 = 两个类之间有方向的连线
- 写清楚：**Domain(起点类) --[关系名]--> Range(终点类)**
- 考虑关系的**基数**：一对一、一对多、多对多

````python
# 🔗 R — 关系建模实践

print("🔗 为大学本体定义关系\n")

relations = [
    ("teaches",       "Teacher",   "Course",    "一对多", "教师教授课程"),
    ("enrolled_in",   "Student",   "Course",    "多对多", "学生选修课程"),
    ("belongs_to",    "Student",   "College",   "多对一", "学生所属学院"),
    ("works_at",      "Teacher",   "Department","多对一", "教师所在系"),
    ("part_of",       "Department","College",   "多对一", "系属于学院"),
    ("supervises",    "Teacher",   "Student",   "一对多", "导师指导学生"),
    ("prerequisite",  "Course",    "Course",    "多对多", "课程先修要求"),
    ("has_grade_in",  "Student",   "Course",    "多对多", "学生在某课的成绩"),
]

print(f"  {'关系名':<16} {'起点(Domain)':<12} {'终点(Range)':<12} {'基数':<8} {'说明'}")
print(f"  {'─'*68}")
for name, domain, range_cls, card, desc in relations:
    print(f"  {name:<16} {domain:<12} {range_cls:<12} {card:<8} {desc}")

# 可视化关系图（ASCII）
print("\n\n📊 关系图（简化版）：")
print("""
  ┌──────────┐    teaches     ┌──────────┐
  │ Teacher  │ ─────────────→ │  Course  │
  └──────────┘                └──────────┘
       │ supervises                ↑ enrolled_in
       ↓                          │
  ┌──────────┐               ┌──────────┐
  │ Student  │ ─────────────→│  Course  │
  └──────────┘  enrolled_in  └──────────┘
       │ belongs_to
       ↓
  ┌──────────┐    part_of    ┌──────────┐
  │ College  │ ←──────────── │Department│
  └──────────┘               └──────────┘
""")
````

### A — Axioms（公理/规则）

> **核心问题：什么是必须的？什么是不允许的？什么可以自动推导？**

**三种公理类型：**
1. **约束型** — "学分必须在 1~6 之间"、"每个订单必须有一个客户"
2. **推理型** — "如果 GPA < 1.5 → 学业预警"
3. **不相容型** — "一个人不能同时是学生和教师"（或者可以？这取决于你的领域）

**🏆 黄金法则：**
- 公理是 Ontology 最有价值的部分——它让机器能"思考"
- 先写业务规则，再翻译成公理
- 公理太多太复杂会降低性能，聚焦关键规则

````python
# ⚖️ A — 公理建模实践

print("⚖️ 为大学本体定义公理\n")

axioms = [
    {
        "类型": "约束型",
        "描述": "学分范围约束",
        "规则": "Course.credits ∈ [1, 6]",
        "自然语言": "每门课的学分必须在 1 到 6 之间"
    },
    {
        "类型": "约束型",
        "描述": "GPA 范围约束",
        "规则": "Student.gpa ∈ [0.0, 4.0]",
        "自然语言": "学生绩点必须在 0 到 4.0 之间"
    },
    {
        "类型": "约束型",
        "描述": "选课必须关联",
        "规则": "enrolled_in(s, c) → s ∈ Student ∧ c ∈ Course",
        "自然语言": "选课关系的两端必须分别是学生和课程"
    },
    {
        "类型": "推理型",
        "描述": "绩点预警推理",
        "规则": "Student(x) ∧ gpa(x) < 1.5 → AcademicWarning(x)",
        "自然语言": "如果学生GPA低于1.5，自动标记学业预警"
    },
    {
        "类型": "推理型",
        "描述": "毕业资格推理",
        "规则": "allRequired(x) ∧ gpa(x) ≥ 2.0 ∧ thesisPass(x) → canGraduate(x)",
        "自然语言": "必修课全部通过 + GPA≥2.0 + 论文通过 → 可以毕业"
    },
    {
        "类型": "推理型",
        "描述": "优秀学生推荐",
        "规则": "Student(x) ∧ gpa(x) ≥ 3.8 → HonorStudent(x)",
        "自然语言": "GPA ≥ 3.8 的学生自动推荐为优秀学生"
    },
    {
        "类型": "不相容",
        "描述": "互斥约束",
        "规则": "Undergraduate ∩ Graduate = ∅",
        "自然语言": "一个学生不能同时是本科生和研究生"
    }
]

for a in axioms:
    icon = {"约束型": "🔒", "推理型": "💡", "不相容": "🚫"}[a["类型"]]
    print(f"  {icon} [{a['类型']}] {a['描述']}")
    print(f"     形式化：{a['规则']}")
    print(f"     含义：{a['自然语言']}")
    print()

print("=" * 60)
print("💡 有了这些公理，Ontology 就能自动：")
print("   1. 拒绝不合法的数据（约束型）")
print("   2. 推导出新的知识（推理型）")
print("   3. 保证逻辑一致性（不相容型）")
````

## 🔄 Part 3：OPERA 五步建模流程

现在你已经理解了每个字母的含义，让我们把它串成一个完整的工作流程：

```
┌─────┐     ┌─────┐     ┌─────┐     ┌─────┐     ┌─────┐
│  O  │ ──→ │  P  │ ──→ │  E  │ ──→ │  R  │ ──→ │  A  │
│对象类型│     │ 属性 │     │事件动作│     │ 关系 │     │ 公理 │
└─────┘     └─────┘     └─────┘     └─────┘     └─────┘
 名词          特征         动词         连接          规则
"有什么？"   "长什么样？"  "能做什么？"  "怎么连？"   "什么规矩？"
```

### 为什么是这个顺序？

1. **先 O** — 不知道有哪些对象，就无法定义属性
2. **再 P** — 知道了对象，才能给它们添加特征
3. **然后 E** — 有了对象和属性，才能定义操作
4. **接着 R** — 有了完整的对象模型，才能描述它们之间的关系
5. **最后 A** — 有了所有的元素，才能写出约束和推理规则

## ✏️ Part 4：动手练习 — 为「外卖平台」建模

> 现在轮到你了！用 OPERA 框架为一个外卖平台（如美团/饿了么）建模。

**场景描述：**
用户通过手机 App 浏览餐厅菜单，下单点餐，骑手接单配送，用户收餐后评价。

**请思考后再运行下面的代码查看参考答案 👇**

提示：
- O：用户、餐厅、菜品、订单、骑手、评价...
- P：每种对象有什么字段？
- E：下单、接单、配送、评价...

````python
# ✏️ 外卖平台 OPERA 建模 — 参考答案

# 先自己思考 30 秒！
# ...
# ...
# 准备好了吗？往下看 ↓

print("🍔 外卖平台 OPERA 建模参考答案")
print("=" * 65)

print("\n📦 O — 对象类型：")
o_types = {
    "User": "用户 (下单的人)",
    "Restaurant": "餐厅",
    "MenuItem": "菜品",
    "Order": "订单",
    "OrderItem": "订单明细（订单中的每一道菜）",
    "Rider": "骑手",
    "Review": "评价",
    "Address": "地址",
}
for name, desc in o_types.items():
    print(f"   · {name} — {desc}")

print("\n📋 P — 属性：")
p_props = {
    "User":       ["user_id(string)", "name(string)", "phone(string)", "vip_level(int)"],
    "Restaurant": ["restaurant_id(string)", "name(string)", "rating(float)", "cuisine(string)", "min_order(float)"],
    "MenuItem":   ["item_id(string)", "name(string)", "price(float)", "available(bool)"],
    "Order":      ["order_id(string)", "total(float)", "status(enum)", "created_at(datetime)", "estimated_delivery(datetime)"],
    "Rider":      ["rider_id(string)", "name(string)", "phone(string)", "current_location(string)", "active(bool)"],
}
for cls, props in p_props.items():
    print(f"   · {cls}: {', '.join(props)}")

print("\n⚡ E — 事件与动作：")
events_actions = [
    "PlaceOrder   — User 下单 → 创建 Order + 通知 Restaurant",
    "AcceptOrder  — Restaurant 接单 → 开始备餐",
    "AssignRider  — 系统分配骑手 → Rider 接到配送任务",
    "PickUp       — Rider 到店取餐 → 更新状态",
    "Deliver      — Rider 送达 → 通知 User 取餐",
    "Review       — User 评价 → 创建 Review + 更新 Restaurant.rating",
    "🔔 OrderTimeout — 当超过预计时间 15 分钟 → 自动退款 + 通知客服",
]
for ea in events_actions:
    print(f"   · {ea}")

print("\n🔗 R — 关系：")
rels = [
    "placed_by:    Order → User          (多对一)",
    "ordered_from: Order → Restaurant    (多对一)",
    "contains:     Order → MenuItem      (多对多，通过 OrderItem)",
    "delivered_by: Order → Rider         (多对一)",
    "belongs_to:   MenuItem → Restaurant (多对一)",
    "reviewed_by:  Review → User         (多对一)",
    "reviews:      Review → Order        (一对一)",
]
for r in rels:
    print(f"   · {r}")

print("\n⚖️ A — 公理：")
axs = [
    "🔒 Order.total ≥ Restaurant.min_order（订单金额不低于起送价）",
    "🔒 MenuItem.available == True 才能加入订单",
    "💡 User 月消费 > 500 → 自动升级 VIP",
    "💡 Restaurant.rating < 3.0 → 标记低分预警",
    "🚫 一个 Order 不能同时状态为 'delivering' 和 'cancelled'",
    "💡 骑手距离餐厅 < 2km 且 active → 优先分配",
]
for a in axs:
    print(f"   · {a}")

print("\n✅ 这就是一个完整的 OPERA 建模！")
````
